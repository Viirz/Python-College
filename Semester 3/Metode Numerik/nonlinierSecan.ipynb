```python
# secan nomer 1

import matplotlib.pyplot as plt
import numpy as np

def secant(f, x0, x1, eps):
    h = f(x1)*(x1-x0)/(f(x1)-f(x0))
    i = 1
    print('n\tx_n0\tx_n1\tf(x_n0)\tf(x_n1)\th')
    print(f'{i:d}\t{x0:.4f}\t{x1:.4f}\t{f(x0):.4f}\t{f(x1):.4f}\t{abs(h):.4f}')

    while abs(h) >= eps:
        x0 = x1
        x1 = x1 - h
        h = f(x1)*(x1-x0)/(f(x1)-f(x0))
        i += 1
        print(f'{i:d}\t{x0:.4f}\t{x1:.4f}\t{f(x0):.4f}\t{f(x1):.4f}\t{abs(h):.4f}')

    print(f'Akar persamaan adalah {x1:.4f}')
    graphic(f, x1)

def graphic(f, x1):
    # Plot the function
    x = np.linspace(0, 2, 1000)
    y = f(x)
    plt.plot(x, y)
    plt.axhline(y=0, color='black', linestyle='--')
    plt.axvline(x=x1, color='red', linestyle='--')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Grafik Fungsi')
    plt.show()

f = lambda x: x**2 * np.exp(-x/2) - 1.0
x0 = 0.0
x1 = 2.0
e = 0.001

secant(f, x0, x1, e)
```
